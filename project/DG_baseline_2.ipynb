```python
class conv(nn.Module):
    def __init__(self,ch_in,ch_out):
        super().__init__()
        self.cell=nn.Sequential(
            nn.Conv2d(ch_in, ch_out, 3, 1, 1),
            nn.BatchNorm2d(ch_out),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.2)
        )
    def forward(self,x):
        return self.cell(x)
    
class DomainClassifier(nn.Module): # input B, 1024, 16*16
    def __init__(self,ch_in=1):
        super().__init__()
        ftrs_ch = 32
        self.blocks = nn.Sequential(
          conv(ch_in,ftrs_ch),
          nn.AdaptiveAvgPool2d((1,1))
        )
        self.output = nn.Sequential(
          nn.Flatten(),
          nn.Linear(ftrs_ch, 128),
          nn.ReLU(),
          nn.Linear(128, 1), 
        )
        
    def forward(self, x):
        x = self.blocks(x)
        x = self.output(x)
        return x
```

## Type 2: Feature extraction, Label prediction and domain confusion
> 把Unet分成Feature extraction, Label prediction兩個部分，這樣我們就可以改成在Unet最底端做domain confusion，當然你也可以透過修改模型(encoder的return)時現任一個位置或多個位置的domain confusion。

```python
from utils.unet import Down, Up, convBlock
```

```python
# self, in_ch, out_ch, bilinear=True, padding=True):
class FeatureExtractor(nn.Module): 
    def __init__(self, out_sz, in_ch, out_ch, bilinear=True, activation=None):
        super().__init__()     
        if isinstance(out_sz,(int)): self.out_sz = (out_sz, out_sz)
        if isinstance(out_sz,(tuple,list)): self.out_sz = tuple(out_sz)
        
        chs = (64, 128, 256, 512, 1024)
        
        '''
        Unet with nn.ModuleList
        '''
        self.input = convBlock(in_ch, chs[0], 3, padding=1)
        self.down_list = nn.ModuleList([Down(chs[i], chs[i+1], padding='same')for i in range(len(chs)-1)]) 
    
    def forward(self, x):
        down_layer_0 = self.input(x)
        
        'Unet with nn.ModuleList'
        enc_ftrs = [down_layer_0]
        for idx in range(len(self.down_list)):
            outputs = self.down_list[idx](enc_ftrs[idx])
            enc_ftrs.append(outputs)
        return enc_ftrs
    
class LabelPredictor(nn.Module):
    def __init__(self, out_sz, in_ch, out_ch, bilinear=True, activation=None):
        super().__init__()
        if isinstance(out_sz,(int)): self.out_sz = (out_sz, out_sz)
        if isinstance(out_sz,(tuple,list)): self.out_sz = tuple(out_sz)
            
        chs = (64, 128, 256, 512, 1024)
        self.head = nn.Conv2d(chs[0], out_ch, 1)
        chs = chs[::-1]
        self.up_list = nn.ModuleList([Up(chs[i], chs[i+1], padding='same')for i in range(len(chs)-1)]) 
        self.activation = activation
        self.out_sz = out_sz
        
    def forward(self, enc_ftrs):
        enc_ftrs = enc_ftrs[::-1]
        
        tmp_ftr = enc_ftrs[0]
        for idx in range(len(self.up_list)):
            tmp_ftr = self.up_list[idx](tmp_ftr, enc_ftrs[idx+1])
            
        logits = self.head(tmp_ftr)
        
        # interpolate 
        _, _, H, W = logits.shape
        if (H,W)==self.out_sz: pass
        else:
            logits = F.interpolate(logits, self.out_sz)
        # add activation (not necessary)
        if self.activation:
            logits = self.activation(logits)
        
        return logits
```

## 建立模型架構

```python
# 建立3個block
feature_extractor = FeatureExtractor((HEIGHT, WIDTH), 1, 1, bilinear=True, activation=None).to(device)
label_predictor = LabelPredictor((HEIGHT, WIDTH), 1, 1, bilinear=True, activation=nn.Sigmoid()).to(device)
domain_classifier = DomainClassifier(ch_in=1024).to(device)
# class_criterion = nn.BCEWithLogitsLoss(pos_weight = torch.Tensor([100]).to(device))  
class_criterion = DiceLoss()
domain_criterion = nn.BCEWithLogitsLoss()
```

```python
import torch.optim as optim
# 為了方便我們令一個class可以直接把dictionary key變成其attribute (可以直接用d.key)
class Dict:
    def __init__(self,d):
        self.__dict__=d
lr=0.01 
optimizers=Dict({name:optim.Adam(method.parameters(),lr=lr) for name,method in 
                (("FE",feature_extractor),
                ("LP",label_predictor),
                ("DC",domain_classifier))
                })
```

- DISC_LOOPS: Discrimination可以比較多次一點，增加Discriminator強度

**Callbacks**
- Learning rate scheduling: ReduceLROnPlateau-模型訓練不下去時learning rate減少


```python
from torchvision.ops import sigmoid_focal_loss
EPOCHS = 1
DISC_LOOPS = 2 # Domain Classifier重複訓練次數
min_target_loss_value = float("inf") 
save_root = './data/save_weights/'
os.makedirs(save_root, exist_ok=True)

# For scheduler
# schedulers=Dict({
#     k:optim.lr_scheduler.ReduceLROnPlateau(v, mode='min', factor=0.1, patience=5)
#     for k,v in optimizers.__dict__.items()
#     }) 
```

### adversarial learning with Unet semantic segmentation

```python
def train_epoch(source_dataloader, target_dataloader, lamb):
    class_loss_value = 0.0 # Domain Classifier的loss和Feature Extrator & Label Predictor的loss
    running_D_loss = 0.0
    for i, ((source_data, source_label), (target_data, target_label)) in enumerate(zip(source_dataloader, target_dataloader)):
        source_data = source_data.to(device)
        source_label = source_label.to(device)
        target_data = target_data.to(device)
        
        target_label = target_label.to(device)
        
        # 我們把source data和target data混在一起，讓normalization標準一致
        mixed_data = torch.cat([source_data, target_data], dim=0)
        # 設定source data的domain label為1, target的為0
        domain_label = torch.zeros([source_data.shape[0] + target_data.shape[0], 1]).to(device)
        domain_label[:source_data.shape[0]] = 1
        
        domain_classifier.train()
        feature_extractor.eval()
        
        with torch.no_grad():
            enc_ftrs = feature_extractor(mixed_data)
        for _ in range(DISC_LOOPS):
            # 不需要訓練Feature Extractor，所以把feature detach避免loss backprop上去。
            domain_logits = domain_classifier(enc_ftrs[-1].detach()) # unet最底下的embedding
            loss = domain_criterion(domain_logits, domain_label)
            running_D_loss+= loss.item()
            loss.backward()
            optimizers.DC.step()
            optimizers.DC.zero_grad()
            del loss, domain_logits
            torch.cuda.empty_cache()
        
        
        feature_extractor.train()
        domain_classifier.eval()
        #訓練Feature Extractor和Label Predictor\
        
        '''
        計算domain fusing loss
        '''
        domain_logits = domain_classifier(enc_ftrs[-1].detach()) # unet最底下的embedding
        loss_disc = domain_criterion(domain_logits, domain_label)
        running_D_loss += loss_disc.item()
        
        '''
        計算特徵萃取器的loss
        '''
        outputs = feature_extractor(source_data)
        outputs = label_predictor(outputs)
        loss = class_criterion(outputs, source_label) + 0.5*lamb*loss_disc
        loss.backward(retain_graph=True)
        
        class_loss_value += loss.item()
        del source_data, source_label, outputs, loss
        torch.cuda.empty_cache()
        
        outputs = feature_extractor(target_data)
        outputs = label_predictor(outputs)
        loss = class_criterion(outputs, target_label) + 0.5*lamb*loss_disc
        loss.backward()
        class_loss_value += loss.item()
        del target_data, target_label, outputs, loss, loss_disc
        torch.cuda.empty_cache()
        

        optimizers.FE.step()
        # schedulers.FE.step(loss) # For scheduler
        optimizers.LP.step()
        # schedulers.LP.step(loss) # For scheduler

        optimizers.DC.zero_grad()        
        optimizers.FE.zero_grad()
        optimizers.LP.zero_grad()
  
        print(i, end='\r')
        del enc_ftrs
        torch.cuda.empty_cache()

    return running_D_loss / (i+1), class_loss_value / (i+1)

# 訓練 epochs

try:
    for epoch in range(EPOCHS):
        train_D_loss, class_loss_value = train_epoch(CT_dataloader_train, dataloader_train, lamb=0.1)
        
        torch.save(feature_extractor.state_dict(), f'{save_root}extractor_model.bin')
        torch.save(label_predictor.state_dict(), f'{save_root}predictor_model.bin')

        print('epoch {:>3d}: DC loss: {:6.4f}, FE+LP loss: {:6.4f}'.format(epoch, train_D_loss, class_loss_value))
except KeyboardInterrupt:
    print("KeyboardInterrupt")
```
